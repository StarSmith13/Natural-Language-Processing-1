```python
"""Take any news article from today’s newspaper and perform POS
tagging onto it to find out how many nouns and how many verbs are there after
removing stop words. Perform lemmatization to the extracted nouns only and
display the result."""
```

```python
import requests
#import beautiful4soup
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Academy_Awards"
```

```python
responses = requests.get(url)
print(responses.content)
```

```python
#soup = BeautifulSoup(responses.content,'html.parser')
#print(soup.prettify())
```

```python
soup = BeautifulSoup(responses.text,'html.parser')
len(responses.text)

paragraphs = soup.find('div', class_='mw-parser-output').find_all('p')
text_body = '\n'.join([paragraph.text.strip() for paragraph in paragraphs])
print(text_body)
```

```python
len(text_body)
```

```python
import nltk
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
article = text_body
```

```python
#Tokenization
tokens = word_tokenize(article)
```

```python
#Remove stopwords
stop_words = set(stopwords.words('english'))
unfiltered_tokens = [word for word in tokens if word.lower() not in stop_words]
filtered_tokens = [word for word in tokens if word.lower() in stop_words]
filtered_tokens
```

```python
# Pos Tagging
pos_tags = nltk.pos_tag(filtered_tokens)
pos_tags
```

```python
type(pos_tags)
```

```python
# Lemmatized
lemmatizer = WordNetLemmatizer()

# Extract nouns and lemmatize them
nouns = [word for word, pos in pos_tags if pos.startswith('N')]
verbs = [word for word, pos in pos_tags if pos.startswith('V')]
lemmatized_nouns = [lemmatizer.lemmatize(word, pos='n') for word in nouns]

# Count nouns and verbs
noun_count = len(nouns)
verb_count = len(verbs)

print("Number of nouns after removing stopwords:", noun_count)
print("Number of verbs after removing stopwords:", verb_count)
print("\nLemmatized nouns:")
print(lemmatized_nouns)
```
